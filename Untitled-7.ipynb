10. Analysis of Different Types of Graphs and Chart Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
data = datasets.load_iris()
```

```python
# Line Plot
sns.lineplot(x=[1, 2, 3], y=[1, 4, 9])
plt.show()
```

```python
# Histogram
sns.histplot(data.data[:, 0], bins=10)
plt.title('Histogram of the first feature')
```

```python
# Bar Plot
sns.barplot(x=['A', 'B', 'C'], y=[3, 7, 5])
plt.show()
```

```python
# Scatter Plot
sns.scatterplot(x=data.data[:, 0], y=data.data[:, 1], hue=data.target)
plt.show()
```
